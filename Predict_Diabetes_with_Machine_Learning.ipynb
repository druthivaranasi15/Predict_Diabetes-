# Are the distributions of numeric variables normal or skewed?

```python
numeric_cols = df.columns[:-1]
n_cols = 3
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols  # auto-adjust rows

plt.figure(figsize=(n_cols * 5, n_rows * 4))

for i, col in enumerate(numeric_cols):
    plt.subplot(n_rows, n_cols, i + 1)
    sns.histplot(df[col], kde=True, bins=30)
    skewness = df[col].skew()
    plt.title(f'{col}\nSkewness: {skewness:.2f}')

plt.tight_layout()
plt.suptitle("Distributions and Skewness of Features", fontsize=16, y=1.03)
plt.show()
```

# Are outliers present in the dataset



```python
for col in ['Insulin', 'SkinThickness']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
for col in ['Insulin', 'SkinThickness']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    print(f'{col} - Number of outliers: {len(outliers)}')
```

Is there a significant difference in BMI between diabetic and non-diabetic groups?

```python
non_diabetic_bmi = df[df['Outcome'] == 0]['BMI']
diabetic_bmi = df[df['Outcome'] == 1]['BMI']

print("Non-Diabetic Mean BMI:", non_diabetic_bmi.mean())
print("Diabetic Mean BMI    :", diabetic_bmi.mean())

print("\nNon-Diabetic Median BMI:", non_diabetic_bmi.median())
print("Diabetic Median BMI    :", diabetic_bmi.median())
```

```python
import plotly.express as px
import pandas as pd

# Custom bin labels
glucose_labels = ["Hypoglycemia", "Healthy Range", "Elevated", "Dangerously High"]
bp_labels = ["Hypotension", "Ideal", "Pre-Hypertension", "Hypertension"]

# Create bins and apply custom labels
df['PregnancyBin'] = pd.cut(df['Pregnancies'], bins=[-1, 0, 2, 5, 10, 20], labels=["None", "1â€“2", "3â€“5", "6â€“10", "10+"])
df['GlucoseLevel'] = pd.cut(df['Glucose'], bins=[0, 90, 120, 150, 200], labels=glucose_labels)
df['BPRange'] = pd.cut(df['BloodPressure'], bins=[0, 60, 80, 100, 150], labels=bp_labels)

# Drop missing rows in path
treemap_df = df.dropna(subset=['PregnancyBin', 'GlucoseLevel', 'BPRange'])

# Create treemap
fig = px.treemap(
    treemap_df,
    path=['PregnancyBin', 'GlucoseLevel', 'BPRange'],
    values=None,
    title="Treemap: Pregnancies â†’ Glucose Level â†’ Blood Pressure"
)

fig.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
training_accuracy=[]
test_accuracy = []
for n_neighbors in range(1,11):
        knn = KNeighborsClassifier(n_neighbors=n_neighbors)
        knn.fit(X_train,y_train)
        training_accuracy.append(knn.score(X_train,y_train))
        test_accuracy.append(knn.score(X_test,y_test))
```

```python
plt.plot(range(1,11), training_accuracy, label= "training_accuracy")
plt.plot(range(1,11),test_accuracy, label="test_accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.show()
```
